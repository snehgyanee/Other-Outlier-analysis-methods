```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('bhp.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.hist(df['price_per_sqft'], bins=20, rwidth=0.8)
plt.xlabel('Price per sqft')
plt.ylabel('Count')
plt.yscale('log')
plt.show()
```

## Standard Deviation method for removing outliers

```python
sd = df['price_per_sqft'].std()
```

```python
sd
```

```python
mean = df['price_per_sqft'].mean()
```

```python
mean
```

```python
upper_limit = mean + 3*sd
```

```python
lower_limit = mean - 3*sd
```

```python
upper_limit,lower_limit
```

```python
df[(df['price_per_sqft']>upper_limit) | (df['price_per_sqft']<lower_limit)]
```

```python
df2 = df[(df['price_per_sqft']<upper_limit) & (df['price_per_sqft']>lower_limit)]
```

```python
df2.head()
```

```python
df2.describe()
```

```python
plt.hist(df2['price_per_sqft'], bins=20, rwidth=0.8)
plt.xlabel('Price per square ft')
plt.ylabel('Count')
plt.yscale('log')
plt.show()
```

## Z score calculation for removing outlier

```python
df['z score'] = (df['price_per_sqft']-df['price_per_sqft'].mean())/df['price_per_sqft'].std()
```

```python
df.head()
```

```python
df.describe()
```

```python
df[(df['z score']<-3) | (df['z score']>3)]
```

```python
data = df[(df['z score']>-3) & (df['z score']<3)]
```

```python
data.head()
```

```python
data.describe()
```

```python
plt.hist(data['price_per_sqft'], bins=20, rwidth=0.8)
plt.xlabel('Price per square ft')
plt.ylabel('Count')
plt.yscale('log')
plt.show()
```
